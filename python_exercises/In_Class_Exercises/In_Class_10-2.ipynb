# Dictionaries

ways to make a series of "key" and "value" pairs

```python
Butler = {'genre':'science fiction', 'award':'hugo'}
```

```python
# how to access information
print(Butler['genre'])
# similar to syntax for lists, but instead of an index number, you input the key to get the value
```

```python
print(Butler['award'])
```

```python
#adding information to a dictionary
Butler['birth year'] = 1947
print(Butler)
```

```python
Robinson = {}
print(type(Robinson))
```

```python
Robinson['genre'] = 'science fiction'
```

```python
print(Robinson)
```

```python
Robinson['award'] = 'hugo'
Robinson['birth year'] = 1952
print(Robinson)
```

## Moving through a dictionary

```python
for key in Butler.keys():
    print(key)
```

```python
for value in Butler.values():
    print(value)
```

```python
for key, value in Butler.items():
    print (key, value)
```

## Word Counter

```python
import requests
```

```python
url = 'https://raw.githubusercontent.com/msaxton/18th-century-historians/main/gibbon/gibbon_decline_and_fall/gibbon_decline_volume1_chap01.txt'
```

```python
response = requests.get(url)
```

```python
text = response.text
```

```python
print(text[:200])
```

```python
word_counts = {}
```

```python
words = text.split()
```

```python
print(words[:100])
```

```python
for word in words:
    lower_word = word.lower()
    if lower_word in word_counts.keys():
        word_counts[lower_word] += 1
    else:
        word_counts[lower_word] = 1
```

```python
for k, v in word_counts.items():
    print(k, v)
```

```python
sorted_word_counts = sorted(word_counts.items(), key=lambda x:x[1], reverse=True)
```

```python
sorted_word_counts
```

```python
# create list of stop words
stop_words = ['the', 'of', 'and', 'to', 'a', 'was', 'in', 'by', 'their', 'were', 'that', 'as', 'which', 'from', 'his', 'with',
             'or', 'it', 'on','they', 'had', 'for', 'but', 'its', 'an', 'is', 'we', 'he', 'than', 'into', 'at', 'be', 'those',
             'this','not', 'who', 'when', 'any', 'very', 'been', 'within', 'nor', 'are', 'these', 'so', 'would', 'which,',
             'without', 'either', 'after', 'about', 'among', 'during', 'before', 'yet',]
# create list of words without stop words
words_no_stops = []
for word in words:  # using original list of words
    word_lower = word.lower()  # we still need to 'normailize' the text by putting everything in lower case
    if word_lower not in stop_words:
        words_no_stops.append(word_lower)
```

```python
word_counts2 = {}
```

```python
for word in words_no_stops:
    word_lower = word.lower()
    if word_lower in word_counts2.keys():
        word_counts2[word_lower] += 1
    else:
        word_counts2[word_lower] = 1
```

```python
sorted_word_counts2 = sorted(word_counts2.items(), key=lambda x:x[1], reverse=True)
```

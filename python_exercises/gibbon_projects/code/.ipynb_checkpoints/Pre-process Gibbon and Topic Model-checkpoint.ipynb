## Prepare Gibbon corpus

```python
# functions
def get_nouns(doc):
    """Return a list of nouns from raw text"""
    tokens = [token for token in doc]
    noun_tokens = [token for token in tokens if token.tag_ == 'NN' or token.tag_ == 'NNP' or token.tag_ == 'NNS']
    noun_lemmas = [noun_token.lemma_ for noun_token in noun_tokens]
    return noun_lemmas

def process_text(text):
    """Use SpaCy for nlp"""
    doc = nlp(text)
    noun_lemmas = get_nouns(doc)
    return noun_lemmas    
```

```python
# create a dictionary of dictionaries
gibbon_dict = {}
for file_name in os.listdir(gibbon_path):
    chapter_name = file_name[23:29]
    gibbon_dict[chapter_name] = {}
    with open(gibbon_path + file_name, encoding='utf-8', mode='r') as f:
        raw_text = f.read()
    gibbon_dict[chapter_name]['raw_text'] = raw_text
```

```python
# sanity check
len(gibbon_dict)
```

```python
print(gibbon_dict.keys())
```

```python
for key in gibbon_dict.keys():
    noun_lemmas = []
    raw_text = gibbon_dict[key]['raw_text']
    text_lines = raw_text.split('\n')  # processing line by line for sake of memory
    for text_line in text_lines:
        nouns = process_text(text_line)
        for noun in nouns:
            noun_lemmas.append(noun)
    gibbon_dict[key]['noun_lemmas'] = noun_lemmas
    
```

## Topic Model

```python
!pip install --upgrade ipykernel --user
```

```python
# imports
from gensim import corpora, models
```

```python
noun_docs = []
for key in gibbon_dict.keys():
    noun_lemmas = gibbon_dict[key]['noun_lemmas']
    noun_docs.append(noun_lemmas)
```

```python
noun_dictionary = corpora.Dictionary(noun_docs)
```

```python
#noun_dictionary.filter_extremes(no_below=1)
```

```python
noun_corpus = [noun_dictionary.doc2bow(doc) for doc in noun_docs]
```

```python
lda_25 = models.LdaModel(noun_corpus, id2word=noun_dictionary, num_topics=25, passes=25, random_state=42)
```
